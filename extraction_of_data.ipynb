```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program


#This is to direct the path to get the data as states

path_trans="pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path_trans)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path_trans+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
```

```python

# Define the path to the directory containing the data
path_user = "pulse/data/aggregated/user/country/india/state/"

# Get the list of states from the directory
Agg_state_list = os.listdir(path_user)

# Initialize a dictionary to collect your data
clu = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [], 'Percentage': []}

# Process each state directory
for i in Agg_state_list:
    state_path = os.path.join(path_user, i)
    year_list = os.listdir(state_path)
    
    for j in year_list:
        year_path = os.path.join(state_path, j)
        quarter_files = os.listdir(year_path)
        
        for k in quarter_files:
            file_path = os.path.join(year_path, k)
            
            # Open and read the JSON file
            try:
                with open(file_path, 'r') as file:
                    data = json.load(file)
                    
                    # Check if 'data' and 'usersByDevice' keys exist
                    if data and "data" in data and "usersByDevice" in data["data"]:
                        for z in data["data"]["usersByDevice"]:
                            clu['State'].append(i)
                            clu['Year'].append(j)
                            clu['Quarter'].append(int(k.strip('.json')))
                            clu['Brands'].append(z['brand'])
                            clu['Count'].append(z['count'])
                            clu['Percentage'].append(z['percentage'])
                    else:
                        print(f"Data missing or 'usersByDevice' is None in file {file_path}")
            except json.JSONDecodeError:
                print(f"JSON decode error in file {file_path}")
            except FileNotFoundError:
                print(f"File not found: {file_path}")
            except Exception as e:
                #print(f"An error occurred while reading {file_path}: {e}")
                continue

# Create a DataFrame from the collected data
Agg_User = pd.DataFrame(clu)
print(Agg_User)
```

```python
path_map = "pulse/data/map/transaction/hover/country/india/state/"
Agg_map_list = os.listdir(path_map)

clmap={'state':[], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for i in Agg_map_list:
    p_i=path_map+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            M=json.load(Data)
            try:
             for z in M['data']['hoverDataList']:
                district=z['name']
                count=z['metric'][0]['count']
                amount=z['metric'][0]['amount']
                clmap['state'].append(i)
                clmap['District'].append(district)
                clmap['Quarter'].append(int(k.strip('.json')))
                clmap['Year'].append(j)
                clmap['Amount'].append(amount)
                clmap['Count'].append(count)
            except json.JSONDecodeError:
                print(f"JSON decode error in file {file_path}")
            except FileNotFoundError:
                print(f"File not found: {file_path}")
            except Exception as e:
                #print(f"An error occurred while reading {file_path}: {e}")
                continue
map_trans=pd.DataFrame(clmap)            
```

```python
path_top = "pulse/data/top/transaction/country/india/state/"
top_trans_list = os.listdir(path_top)
cltop = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for i in top_trans_list:
    cur_state = path_top + i + "/"
    top_year_list = os.listdir(cur_state)
    
    for j in top_year_list:
        cur_year = cur_state + j + "/"
        top_file_list = os.listdir(cur_year)
        
        for k in top_file_list:
            cur_file = cur_year + k
            data = open(cur_file, 'r')
            top = json.load(data)
            
            for z in top['data']['pincodes']:
                name = z['entityName']
                count = z['metric']['count']
                amount = z['metric']['amount']
                cltop['Pincode'].append(name)
                cltop['Transaction_count'].append(count)
                cltop['Transaction_amount'].append(amount)
                cltop['State'].append(i)
                cltop['Year'].append(j)
                cltop['Quarter'].append(int(k.strip('.json')))
df_top_trans = pd.DataFrame(cltop)
```

```python
path_tuser = "pulse/data/top/user/country/india/state/"
top_user_list = os.listdir(path_tuser)
cltu = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [],
            'RegisteredUsers': []}

for i in top_user_list:
    cur_state = path_tuser + i+ "/"
    top_year_list = os.listdir(cur_state)
    
    for j in top_year_list:
        cur_year = cur_state + j + "/"
        top_file_list = os.listdir(cur_year)
        
        for k in top_file_list:
            cur_file = cur_year + k
            data = open(cur_file, 'r')
            F = json.load(data)
            
            for z in F['data']['pincodes']:
                name = z['name']
                registeredUsers = z['registeredUsers']
                cltu['Pincode'].append(name)
                cltu['RegisteredUsers'].append(registeredUsers)
                cltu['State'].append(i)
                cltu['Year'].append(j)
                cltu['Quarter'].append(int(k.strip('.json')))
df_top_user = pd.DataFrame(cltu)
```
